# The Cost of Living Crisis: A Data-Driven Analysis

## The Problem

Official inflation metrics, such as the U.S. Consumer Price Index (CPI), are designed to reflect the spending patterns of the average urban consumer. However, this national average can obscure the lived reality of specific groups. For students, whose budgets are disproportionately allocated toward tuition and housing, headline CPI figures may understate the true increase in the cost of living.

This issue is further amplified in high-cost metropolitan areas such as Boston, where regional inflation consistently exceeds national trends. As a result, students face a compounded inflation burden that is not fully captured by standard CPI measures.

---

## Methodology

This project constructs a custom **Student-Specific Price Index (Student CPI)** using Python and publicly available data from the Federal Reserve Economic Data (FRED) API.

Key methodological steps include:

- Collecting CPI sub-indices for tuition, rent, food away from home, streaming services, and transportation as proxies for student consumption.
- Re-indexing all CPI series to a common base period (**January 2016 = 100**) to ensure comparability across categories with different base years.
- Applying a **Laspeyres-style index framework** with student-relevant expenditure weights, emphasizing tuition and housing costs.
- Comparing the resulting Student CPI against both the official national CPI and the Boston-Cambridge-Newton regional CPI.

This approach enables a transparent comparison between aggregate inflation metrics and the inflation actually experienced by students.

---

## Key Findings

The analysis reveals a clear divergence between student-specific inflation and official CPI measures in the post-2016 period. While national CPI suggests moderate inflation, the Student CPI rises more rapidly due to the outsized contribution of tuition and housing costs.

By the end of the sample period, the Student CPI exceeds the national CPI by approximately **2–3%**, indicating that student living costs have grown meaningfully faster than what is implied by headline inflation statistics. Additionally, the Boston-area CPI consistently tracks above the national average, further intensifying the cost-of-living pressures faced by students in the region.

---

## Conclusion

These findings highlight a key limitation of relying on aggregate inflation metrics to assess economic well-being. Inflation is not experienced uniformly; it varies by geography and by expenditure composition. For students—particularly those living in high-cost cities like Boston—the true cost of living crisis is understated by national averages.

This project demonstrates how disaggregated data, index theory, and transparent weighting schemes can be used to construct more representative measures of economic reality.
