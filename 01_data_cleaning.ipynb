**Data Ingestion Note (Raw Reality) **

Unlike clean CSV inputs, the Angrist & Evans (1998) replication data are provided as SAS datasets (`.sas7bdat`).  
Therefore, I use `pandas.read_sas()` to ingest the raw files. Some columns are byte-encoded and must be decoded before numeric conversion.

```python
import zipfile
import os

zip_path = "/content/AngEv98.zip"
extract_path = "AngEv98"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Unzipped files:")
print(os.listdir(extract_path))
```

```python
import pandas as pd
```

```python
df_1980 = pd.read_sas("AngEv98/AngEv98/m_d_806.sas7bdat")
df_1990 = pd.read_sas("AngEv98/AngEv98/m_d_903.sas7bdat")
```
