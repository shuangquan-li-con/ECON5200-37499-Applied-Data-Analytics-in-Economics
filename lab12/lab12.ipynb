```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from statsmodels.tools.eval_measures import rmse
import matplotlib.pyplot as plt

# Step 1: Ingestion from external source
url = 'https://raw.githubusercontent.com/shuangquan-li-con/ECON5200-37499-Applied-Data-Analytics-in-Economics/refs/heads/main/data/Zillow_ZHVI_2026_Micro.csv'
df = pd.read_csv(url)
```

```python
formula = 'Home_Value ~ Property_Age +Distance_to_Transit +School_District_Rating'
```

```python
model = smf.ols(formula=formula,data=df)
results = model.fit()
print(results.summary())
```

```python
y_pred = results.predict(df)
model_rmse = rmse(df['Home_Value'],y_pred)
print(f"\nThe Predictive RMSE is: ${model_rmse:,.2f}")
```

## Prepare Data for Plotly

### Subtask:
Extract the residuals and fitted values from the 'results' object. Calculate the standard deviation of the residuals and create a boolean column to identify outliers that exceed 2 standard deviations for highlighting in the plot.


**Reasoning**:
The subtask requires extracting residuals and fitted values from the `results` object, calculating the standard deviation of residuals, and then identifying outliers based on a threshold of two standard deviations. I will write a code block to perform these calculations and create the necessary series.



```python
residuals = results.resid
fitted_values = results.fittedvalues
std_residuals = residuals.std()
is_outlier = abs(residuals) > 2 * std_residuals

print(f"Number of outliers based on 2 standard deviations: {is_outlier.sum()}")
print(f"Standard deviation of residuals: {std_residuals:,.2f}")
```

**Reasoning**:
The subtask now requires generating an interactive Plotly scatter plot using the extracted residuals and fitted values. I need to import Plotly Express, create the scatter plot, differentiate outliers, add a zero-line, and set appropriate labels and a title.



```python
import plotly.express as px

fig = px.scatter(
    x=fitted_values,
    y=residuals,
    color=is_outlier,
    labels={
        "x": "Fitted Values",
        "y": "Residuals",
        "color": "Is Outlier (abs(residual) > 2*std_residuals)"
    },
    title="Residual Plot (Residuals vs. Fitted Values)"
)

fig.add_hline(y=0, line_dash="dash", line_color="red", annotation_text="Zero Residual Line", annotation_position="bottom right")

fig.show()
```

### Interpretation of the Residual Plot

The residual plot displays the residuals (errors) of the regression model against the fitted (predicted) values. This visualization is crucial for diagnosing potential issues such as heteroscedasticity or structural breaks in the model.

**Key observations and their implications:**

*   **Random Scatter around Zero:** Ideally, for a well-fitted model, the residuals should be randomly scattered around the zero horizontal line, with no discernible pattern. This indicates that the model's assumptions (e.g., constant variance of errors) are likely met.

*   **Heteroscedasticity:** If the spread (variance) of the residuals changes as the fitted values increase or decrease (e.g., forming a cone or fan shape), this suggests heteroscedasticity. Heteroscedasticity violates the assumption of constant variance and can lead to inefficient coefficient estimates and incorrect standard errors, affecting the reliability of hypothesis tests.

*   **Structural Breaks / Non-linearity:** If the residuals show a distinct pattern (e.g., a curve, or a sudden shift), it might indicate that the linear model is not appropriate, or there are structural breaks in the data that the model is not capturing. This suggests non-linearity in the relationship between predictors and the response variable, or missing important variables.

*   **Outliers:** The points highlighted in a different color (identified as `is_outlier` in the plot) represent observations where the absolute residual value is greater than two standard deviations. These outliers can disproportionately influence the regression line and may warrant further investigation. They could be data entry errors, unusual observations, or indicators of a poor model fit for specific data points.
