```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import IsolationForest
```

```python
from pickle import TRUE
data=fetch_california_housing(as_frame=True)
df=data.frame
```

```python
print(df['MedHouseVal'].describe())
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['MedHouseVal'],bins=50,kde=True)
plt.title("Distribution of House Values: The $500k Cap")
plt.xlabel("Median House Value ($100k)")
plt.show()
```

```python
def flag_outliers_iqr(df,column):
  Q1=df[column].quantile(0.25)
  Q3=df[column].quantile(0.75)
  IQR=Q3-Q1
  lower_bound=Q1-1.5*IQR
  upper_bound=Q3+1.5*IQR
  return (df[column] < lower_bound)|(df[column]>upper_bound)
```

```python
df['outlier_iqr']=flag_outliers_iqr(df,'MedInc')
```

```python
print(f"Manual Outliers Detected:{df['outlier_iqr'].sum()}")
print("These are primarily the 'Wealthy' districts.")
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(x=df['MedInc'],color='lightblue')
plt.title('Boxplot of Median Income (Detecting the "Elon Musk" Effect)')
plt.show()
```

```python
iso_forest = IsolationForest(n_estimators=100,contamination=0.05,random_state=42)
```

```python
features =['MedInc','HouseAge','AveRooms','AveBedrms','Population']
```

```python
preds=iso_forest.fit_predict(df[features])
```

```python
df['outlier_iso']=preds==-1
```

```python
print(f"Algorithmic Outliers Detected:{df['outlier_iso'].sum()}")
```

```python
plt.figure(figsize=(12,6))
```

```python
sns.scatterplot(
    data=df,
    x='MedInc',
    y='AveRooms',
    hue='outlier_iso',
    palette={True:'red',False:'blue'},
    alpha=0.6
)
plt.title("Isolation Forest Anomaly Detection: Income vs. Rooms")
plt.xlabel("Median Income")
plt.ylabel("Average Rooms")
plt.ylim(0,20)
plt.legend(title='Is Anomaly?')
plt.show()
```

## Split DataFrames

### Subtask:
Split the main DataFrame 'df' into 'df_normal' and 'df_outlier' based on the 'outlier_iso' column.


**Reasoning**:
I will create two new DataFrames, `df_normal` and `df_outlier`, by filtering the main DataFrame `df` based on the 'outlier_iso' column, as per the subtask instructions.



```python
df_normal = df[df['outlier_iso'] == False]
df_outlier = df[df['outlier_iso'] == True]

print(f"Number of normal instances: {len(df_normal)}")
print(f"Number of outlier instances: {len(df_outlier)}")
```

**Reasoning**:
Now that the dataframes are split, I will calculate the descriptive statistics (mean, median, standard deviation, and Median Absolute Deviation (MAD)) for 'MedInc' and 'MedHouseVal' for both the normal and outlier groups. I'll also calculate the 'Inequality Wedge' (Mean - Median) for the outlier group, as per the main task.



```python
from statsmodels.robust import mad

def calculate_descriptive_stats(df, column_name, group_name):
    mean_val = df[column_name].mean()
    median_val = df[column_name].median()
    std_val = df[column_name].std()
    mad_val = mad(df[column_name])

    stats = {
        'Group': group_name,
        'Metric': column_name,
        'Mean': mean_val,
        'Median': median_val,
        'Std Dev': std_val,
        'MAD': mad_val
    }

    if group_name == 'Outlier':
        stats['Inequality Wedge (Mean - Median)'] = mean_val - median_val

    return stats

metrics = ['MedInc', 'MedHouseVal']
all_stats = []

for metric in metrics:
    all_stats.append(calculate_descriptive_stats(df_normal, metric, 'Normal'))
    all_stats.append(calculate_descriptive_stats(df_outlier, metric, 'Outlier'))

summary_df = pd.DataFrame(all_stats)
print("Summary of Descriptive Statistics for Normal and Outlier Groups:")
print(summary_df.to_string(index=False))
```

**Reasoning**:
I need to create a 1x2 subplot showing histograms of 'MedInc' for both normal and outlier groups, as specified in the main task. This will visually compare the distribution of median income in both groups.



```python
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
sns.histplot(df_normal['MedInc'], bins=50, kde=True, color='blue')
plt.title('Distribution of Median Income (Normal)')
plt.xlabel('Median Income')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.histplot(df_outlier['MedInc'], bins=50, kde=True, color='red')
plt.title('Distribution of Median Income (Outliers)')
plt.xlabel('Median Income')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
